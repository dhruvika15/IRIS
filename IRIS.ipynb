```python
df['label'] = iris.target
```

```python
y = df['label']
```

```python
df.columns
```

```python
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
```

```python
X
```

```python
y
```

```python
 #One hot encoding
y_final = pandas.get_dummies(y)
```

```python
y_final
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_final, test_size=0.10, random_state=42)
#random_state shift+tab
```

```python
from keras import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import Dense
'''
model.add([
    Dense()
])'''
```

```python
model = Sequential(
    [ 
        Dense(  units=32, activation='relu', input_shape=(4,)  ),
        Dense(units=32, activation='relu'), 
        Dense(units=16, activation='relu'),
        Dense(units=16, activation='relu'),
        
        Dense(units=8, activation='relu'),
        
        Dense(units=8, activation='relu'),
        Dense(units=3, activation='softmax'),
    ]
)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics= ['accuracy'])
```

```python
history= model.fit(X_train, y_train, epochs=20, validation_split= 0.20, batch_size= 20)
```
